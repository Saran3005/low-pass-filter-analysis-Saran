### 9. Recursive Function with Conditional Twist

What is returned by mystery(6)?
```py
def mystery(n):
    if n <= 1:
        return 1
    if n % 2 == 0:
        return n * mystery(n-1)
    return mystery(n-1) + 1
```

````python
def mystery(n):
    if n <= 1:
        return 1
    if n % 2 == 0:
        return n * mystery(n-1)
    return mystery(n-1) + 1

mystery(6)
````

mystery(2) = 2 * 1 = 2
mystery(3) = mystery(2) + 1 = 2 + 1 = 3
mystery(4) = 4 * 3 = 12
mystery(5) = 12 + 1 = 13
mystery(6) = 6 * 13 = 78
